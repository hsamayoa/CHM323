# Linear regression

To perform a linear reression we will use the linregress function from the scipy library.

First, we will define our x and y arrays in the following cell. Feel free to modify the x and y arrays, you need to run the cell to load them.

```python
x=np.array([0,0.1,0.2,0.4,0.8])       #Concentration of riboflavin [Î¼g/mL]
y=np.array([0,5.8,12.2,22.3,43.3])    #Fluorescence intensity [arb.units]
```

The next code cell does not need modification. There are some comments next to each line to help you understand what the code is doing. Read them and make sense of them.

After running the code cell, you will see important the statistical output of the linear regression.

*Slope* is the slope of the linear regression.
*Intercept

```python
from scipy import stats                                #This line imports stats from the scipy library
slope,intercept,r,prob2,sm=stats.linregress(x,y)    #This line indicates that the output will be an array of values: slope (m), intercept (b), correlation coefficient (r),
                                                #p-value for H0 being slope =0 (prob2) and, standard uncertainty of the slope(m).

x_av=x.mean()                                        #Calculates the mean of the x values used for the regression
sx2=((x_av-x)**2).sum()                              #Calculates the sum of the squares of (x_average - x)
sy=(sm**2*sx2)**0.5                                  #Calculates the standard deviation of the predicted y-value
sb=sy*1/(len(x)-x.sum()**2/(x**2).sum())**0.5        #Calculates the standard uncertainty of the intercept

#The next lines print the results of the linear regression. Note that text to be printed is writen between '' and followed by a comma.
print('slope=',round(slope,4))                        
print('intercept=',round(intercept,6))
print('R^2=',round(r**2,4))
print('sy=',round(sy,4))
print('sm=',round(sm,4))
print('sb=',round(sb,4))
```

Great. Now that we know the results of the regression we can draw the line of best-fit for our data and compare it with our experimental data.

```python
fit_ydata=slope*x+intercept                                     #This line uses the intercept and slope calculated from the previous code cell
                                                                #to calculate predicted y-values.

#Plot line of best fit with experimental data
plt.figure(figsize=(8,6))                           
plt.rc('font', size=16)                             
plt.scatter(x,y,color='g',marker='v')                              #This command will plot the experimental data as scatter plot.                     
plt.xlabel('Riboflavin concentration [$\mu$g/mL]',fontsize='16')   
plt.ylabel('Fluorescence intensity [a.u.]',fontsize='16')         
plt.title('Riboflavin calibration plot',fontsize='16')       
plt.plot(x,fit_ydata, color='g',linestyle='--')                    #Plots the line of best fit.
print('y=', round(slope,4),'*x +',round(intercept,6))              #Prints the equation of the line
```

Another important parameter to analyze is the residual. Residuals are the difference between the observed y-value and the predicted y-value from the linear regression. You can see a residual as an error. In general, residuals show be small numbers. If a residual is plot against concentration we obtain a residual plot. This graph helps us visualize how residuals distribute in our calibration plot. Ideally, the residuals should be randomnly distributed around zero.

```python
residual=y-fit_ydata

#Plot residuals
plt.figure(figsize=(8,6))                           
plt.rc('font', size=16)                             
plt.scatter(x,residual)                                
plt.xlabel('Riboflavin concentration [$\mu$g/mL]',fontsize='16')   
plt.ylabel('Residual [a.u.]',fontsize='16')         
plt.title('Residual plot',fontsize='16')       
```
